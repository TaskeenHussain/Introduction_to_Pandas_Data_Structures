### Basic_Functionality_pandas_series

```python
import pandas as pd
import numpy as np

# Creating a Series
# A Series is a one-dimensional labeled array that holds data of any type.
s = pd.Series(np.random.randn(4))
print(s)

# Axes
# Returns the row axis labels of the Series.
print("The axes are:")
print(s.axes)

# Empty
# Checks if the Series is empty.
print("Is the Object empty?")
print(s.empty)

# Dimensions
# Returns the number of dimensions of the Series.
print("The dimensions of the object:")
print(s.ndim)

# Size
# Returns the number of elements in the Series.
print("The size of the object:")
print(s.size)

# Values
# Returns the actual data in the Series as a NumPy array.
print("The actual data series is:")
print(s.values)

# Head & Tail
# Displays the first and last n rows of the Series.
print("The first two rows of the data series:")
print(s.head(2))
print("The last two rows of the data series:")
print(s.tail(2))

# DataFrame Basic Functionality
# Creating a DataFrame from a dictionary of Series.
d = {
    'Name': pd.Series(['Tom', 'James', 'Ricky', 'Vin', 'Steve', 'Smith', 'Jack']),
    'Age': pd.Series([25, 26, 25, 23, 30, 29, 23]),
    'Rating': pd.Series([4.23, 3.24, 3.98, 2.56, 3.20, 4.6, 3.8])
}
df = pd.DataFrame(d)
print("Our data frame is:")
print(df)

# Transpose
# Swaps rows and columns.
print("The transpose of the data frame is:")
print(df.T)

# Axes
# Returns row and column labels of the DataFrame.
print("Row axis labels and column axis labels are:")
print(df.axes)

# Data Types
# Returns the data type of each column.
print("The data types of each column are:")
print(df.dtypes)

# Empty Check
# Checks if the DataFrame is empty.
print("Is the DataFrame empty?")
print(df.empty)

# Number of Dimensions
# Returns the number of dimensions of the DataFrame.
print("The dimension of the object is:")
print(df.ndim)

# Shape
# Returns the number of rows and columns in the DataFrame.
print("The shape of the object is:")
print(df.shape)

# Size
# Returns the total number of elements in the DataFrame.
print("The total number of elements in our object is:")
print(df.size)

# Values
# Returns the underlying data of the DataFrame as a NumPy array.
print("The actual data in our data frame is:")
print(df.values)

# Head & Tail
# Displays the first and last n rows of the DataFrame.
print("The first two rows of the data frame are:")
print(df.head(2))
print("The last two rows of the data frame are:")
print(df.tail(2))
```

### Code to create a Pandas Series with 100 random numbers:

####  Explanation:
##### np.random.randn(100) #- Generates 100 random numbers from a standard normal distribution (mean=0, std=1)
##### pd.Series() #- Converts the NumPy array to a Pandas Series

```python

import pandas as pd
import numpy as np

# Create a series with 100 random numbers (standard normal distribution)
s = pd.Series(np.random.randn(100))
print(s)
```
