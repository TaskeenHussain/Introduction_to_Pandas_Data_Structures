```python
###  Basic Plotting: Line Plot

import pandas as pd
import numpy as np

# Create a DataFrame with random data
df = pd.DataFrame(np.random.randn(10, 4), index=pd.date_range('1/1/2000', periods=10), columns=list('ABCD'))

# Plot a line chart
df.plot()
##This creates a line plot of random data (randn) over 10 days, with columns labeled as A, B, C, D.

```

```python
## 2. Bar Plot
import pandas as pd
import numpy as np

# Create a DataFrame with random data
df = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])

# Plot a bar chart
df.plot.bar()

### A bar plot is created using the plot.bar() method. Each bar represents the values for one of the columns (a, b, c, d).
```

```python
### 3. Stacked Bar Plot

import pandas as pd
import numpy as np

# Create a DataFrame with random data
df = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])

# Plot a stacked bar chart
df.plot.bar(stacked=True)

## Stacked bar plots show the cumulative values of each column in one bar
```

```python
### 4. Horizontal Stacked Bar Plot

import pandas as pd
import numpy as np

# Create a DataFrame with random data
df = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])

# Plot a horizontal stacked bar chart
df.plot.barh(stacked=True)

### Similar to the stacked bar plot but displayed horizontally.
```

```python
### 5. Histograms

import pandas as pd
import numpy as np

# Create a DataFrame with random data
df = pd.DataFrame({'a': np.random.randn(1000) + 1, 'b': np.random.randn(1000), 'c': np.random.randn(1000) - 1})

# Plot a histogram
df.plot.hist(bins=20)

### A histogram displays the distribution of numerical data, with 20 bins representing data intervals.
```

```python
### 6. Individual Histograms
import pandas as pd
import numpy as np

# Create a DataFrame with random data
df = pd.DataFrame({'a': np.random.randn(1000) + 1, 'b': np.random.randn(1000), 'c': np.random.randn(1000) - 1, 'd': np.random.randn(1000) - 2})

# Plot individual histograms
df.hist(bins=20)

### Each column is plotted in its own histogram, showing the data distribution independently.
```

```python
##  Box Plot

import pandas as pd
import numpy as np

# Create a DataFrame with random data
df = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])

# Plot a box plot
df.plot.box()

### Box plots show the distribution, median, and outliers for each column in the DataFrame.

```

```python
### 8. Area Plot
import pandas as pd
import numpy as np

# Create a DataFrame with random data
df = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])

# Plot an area chart
df.plot.area()

### Area plots show the cumulative values of columns over an index.
```

```python
## 9. Scatter Plot
import pandas as pd
import numpy as np

# Create a DataFrame with random data
df = pd.DataFrame(np.random.rand(50, 4), columns=['a', 'b', 'c', 'd'])

# Plot a scatter plot
df.plot.scatter(x='a', y='b')

### A scatter plot shows the relationship between two variables, a (x-axis) and b (y-axis).
```

```python
###10. Pie Chart
import pandas as pd
import numpy as np

# Create a DataFrame with random data
df = pd.DataFrame(3 * np.random.rand(4), index=['a', 'b', 'c', 'd'], columns=['x'])

# Plot a pie chart
df.plot.pie(subplots=True)

###A pie chart is created for each column, showing proportional values of data.

```
